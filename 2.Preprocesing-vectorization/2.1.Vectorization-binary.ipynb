```python
# Librerias
import pandas as pd
import numpy as np
from pprint import pprint
import plotly.graph_objects as go
```

```python
#Funcion para leer Excel

def leerExcel(año):
    dataAño = pd.read_excel("Data-resoluciones.xlsx",
                             sheet_name=año,
                             usecols=[0,1],
                             header=None)
    return dataAño
```

```python
#Concatenando valores de Excel de cada año
DataTotal = pd.concat([leerExcel("2006"), leerExcel("2007"), 
                       leerExcel("2008"), leerExcel("2009"),
                       leerExcel("2010"), leerExcel("2011"), leerExcel("2012"), 
                       leerExcel("2013"), leerExcel("2014"),
                       leerExcel("2015"), leerExcel("2016"), leerExcel("2017"),
                        leerExcel("2018"), leerExcel("2019"), leerExcel("2020")], 
                      ignore_index=True)
```

```python
## Creando un diccionario que convierta a lista las normativas (para luego extraerlas)

resoluciones = {}
num_res = -1
variables_listas = ['Número de resolución', 'Normativa',
                    "Configuración del Tribunal",
                    "Empresa denunciada",
                    "Decisión"]

filas = len(DataTotal)
variables = list( DataTotal.iloc[:,0].value_counts().index )

for fila in range(filas):
    # Cada resolución será un diccionario dentro del diccionario

    # Debe crearse una resolución si es la primera fila o si la última fila revisada fue "Decisión"
    if fila == 0 or DataTotal.iloc[fila-1,0] == 'Decisión':
        num_res += 1
        resoluciones[num_res] = {}
        
    # Si en la primera fila hay una variable, creamos una llave en el diccionario de la resolución
    if DataTotal.iloc[fila, 0] in variables:
        variable = DataTotal.iloc[fila, 0]
        if variable in variables_listas:
            resoluciones[num_res][variable] = []
        else:
            resoluciones[num_res][variable] = ""

    # En las otras dos columnas, si el valor no es un vacío se agregará a la última variable registrada como llave
    for columna in range(1, 2):
        celda = DataTotal.iloc[fila, columna]
        if pd.notna( celda ):
            # Si el valor de la variable llave es vacío, no necesita agregarle espacios
            if variable in variables_listas:
                resoluciones[num_res][variable].append(celda)
            else:
                if resoluciones[num_res][variable] == "":
                    resoluciones[num_res][variable] += str(celda)
                else:
                    resoluciones[num_res][variable] += " " + str(celda)

pprint(resoluciones)
```

```python
## Extrayendo las normativas con el ID de resoluciones y decisiones (esto luego nos permitira hacer Arbol de dec)

normativas = ['Número de resolución', 'Normativa',
              "Empresa denunciada", 'Configuración del Tribunal']

dictNormativas = {}
for resolucion in resoluciones:
    dictNormativas[resolucion] = {}
    for variable in resoluciones[resolucion]:
        if variable in normativas:
            dictNormativas[resolucion][variable] = resoluciones[resolucion][variable]
            
pprint(dictNormativas)
```

```python
dictNormativas_serie = pd.Series(dictNormativas)
```

```python
#Trayendo solo las normas para analizarlas

normativas = ['Normativa']
normas = []
for resolucion in dictNormativas:
    for variable in dictNormativas[resolucion]:
        if variable in normativas:
            lista = dictNormativas[resolucion][variable]
            normas += lista

normas = pd.Series(normas)
```

```python
#contando ocurrencias

countNormas = normas.value_counts()
```

```python
#convirtiendo a DF las normas y sus ocurrencias

DFNormas = pd.DataFrame(countNormas)
```

```python
#Exportando a csv para validarlo

DFNormas.to_csv("mp-nomr.csv")
```
